```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
np.random.seed(0)
num_points = 100
a = 4.31
b = -4.31
noise = np.random.normal(0, 1, num_points)
x = np.random.uniform(0, 5, num_points)
y = a * x + b + noise

plt.scatter(x, y, marker = '.', color = 'r')
plt.show()
#plt.plot(x, a * x + b)
```

```python
#Gradient Descent
def gradient_descent():
    a = 0
    b = 0
    num_points = 1000
    rate = 0.1
    n = float(len(x))
    for i in range (num_points):
        y_new = a*x + b
        dA =(-2/n) * sum(x * (y - y_new))
        dB =(-2/n) * sum(y - y_new)
        a = a - rate * dA
        b = b - rate * dB

    print (a, b)

    y_new = a*x + b
    plt.scatter(x, y, marker = '.', color = 'r')
    plt.plot([min(x), max(x)], [min(y_new), max(y_new)], color='b')
    plt.show()
#plt.plot(x, a * x + b)

gradient_descent()
```
